## Step 5 - Determine a suitable way to combine the orders_products_combined dataframe with your products data set. Make sure you’re using your wrangled, cleaned, and deduped products data set stored in your “Prepared Data” folder from the previous Exercise’s task.

```python
#importing the wrangled products df
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_cleaned_final.csv'), index_col = 0)
```

```python
#viewing df prods
df_prods.head()
```

```python
#viewing df_orders_products_combined 
df_orders_products_combined.head()
```

```python
#dropping merge flag from above df
df_orders_products_combined= df_orders_products_combined.drop(columns=['_merge'])
```

```python
#merging the previous two dfs
df_orders_products_merged = df_orders_products_combined.merge(df_prods, on = 'product_id',indicator= True)
```
